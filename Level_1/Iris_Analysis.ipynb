```python
# STAGE 1 : Import the tools
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

# STAGE 2 : Connect to Google Drive
drive.mount('/content/drive')

# STAGE 3 :Dataset Loading :
path = '/content/drive/MyDrive/Codveda_Internship/iris.csv'
df = pd.read_csv(path)

# STAGE 4 : Verification  (Task 1)
print("Dataset displaying :")
display(df.head())
```

```python
# --- Summary Statistics (Task 2) ---
print("Here are the averages for your flowers :")
print(df.select_dtypes(include=['number']).mean())

# --- GRAPHIC DRAWING (Task 2) ---
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.countplot(x='species', data=df)
plt.title("Number of flowers per species")
plt.show()
```

```python
# A stunning graphic for the final video
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.title("Relationship between petal length and width")
plt.show()
```
